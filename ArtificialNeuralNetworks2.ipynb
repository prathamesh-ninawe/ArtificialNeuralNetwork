### Neural Network Architecture

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense

y = df['TEY']
X = df.drop("TEY", axis=1)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state =2)

from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.fit_transform(X_test)
```

```python
model = Sequential()
model.add(Dense(36, input_dim = 10, activation = 'relu')) #i
model.add(Dense(18, activation = 'relu')) #h1
model.add(Dense(9,activation = 'relu')) #h2
model.add(Dense(1, activation = 'linear')) #o
```

```python
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_absolute_error'])
```

```python
history = model.fit(X_train, y_train, epochs=300, batch_size=200, validation_split=0.123)
```

```python
model.summary()
```

```python
pred_y = model.predict(X_test)
pred_y
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,pred_y)
```
